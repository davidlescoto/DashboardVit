```python
PATH_DATA = "https://raw.githubusercontent.com/davidlescoto/data_sets/main/vitesse_dataset.csv"

#----------------CLEAN DATA----------------------------------------
df = pd.read_csv(PATH_DATA)
dff = df.loc[~((df['poutcome'] == 'nonexistent') & (df['y'] =="yes"))]
dff = dff.drop_duplicates()
dff = dff[dff.marital != "unknown"]
```

```python
import plotly.express as px
import plotly.figure_factory as ff
```

```python
age = dff.age.to_list()
fig_age = ff.create_distplot([age], ["age"],
                         bin_size=.2, show_rug=False)

# Add title
fig_age.update_layout(title_text='Distribucin por edad')
fig_age.show()
```

```python
campaign = dff.campaign.to_list()
fig_campaign = ff.create_distplot([campaign], ["Campaing"],
                         bin_size=.2, show_rug=False)

# Add title
fig_campaign.update_layout(title_text='Distribucin por campaa')
fig_campaign.show()
```

```python
def bar_group(dff, x, y):
    x = "marital"
    y = "y"
    y_values = dff[y].unique()
    data = []
    for value in y_values:
        dfff = dff[dff[y]==value]
        dfff = dfff.groupby([x]).count().reset_index()
        xx = dfff[x].to_list()
        yy = dfff[y].to_list()
        data.append(go.Bar(name = value, x = xx ,y = yy ))


    fig = go.Figure(data=data)
    # Change the bar mode
    fig.update_layout(barmode='group')
    return fig
```

```python
bar_group(dff, "marital", "y")
```

```python
#quantiles_tasa
def cohorte_numericas(dff, column):
    
    quantil = dff[column].quantile(0)
    primer_quantil = dff[column].quantile(.20) # Se toma el valor del primer cuantil
    segundo_quantil = dff[column].quantile(.40)
    tercer_quantil = dff[column].quantile(.60)
    cuarto_quantil = dff[column].quantile(.80)
    quinto_quantil = dff[column].quantile(1)
    quantiles = [quantil,primer_quantil,segundo_quantil,tercer_quantil,cuarto_quantil, quinto_quantil]
    grp_dff = dff.groupby([pd.cut(dff[column], quantiles), "y"])[["y"]].count()
    grp_dff  = grp_dff.rename(columns = {"y":"Total"}).reset_index()

    return grp_dff



```

```python
euro = cohorte_numericas(dff, "euribor3m")
```

```python
def bar_group_cohor(dff, x, y, values):

    y_values = dff[y].unique()
    data = []
    
    for value in y_values:
        dfff = dff[dff[y]==value]
        
        xx = [str(index) for index in dfff[x].to_list()]
        yy = dfff[values].to_list()
        data.append(go.Bar(name = value, x = xx ,y = yy ))


    fig = go.Figure(data=data)
    # Change the bar mode
    fig.update_layout(barmode='group')
    return fig
```

```python
bar_group_cohor(euro, x = "euribor3m", y = "y", values = "Total")
```

```python
price_index = cohorte_numericas(dff, "cons.price.idx")
bar_group_cohor(price_index, x = "cons.price.idx", y = "y", values = "Total")
```

```python
var_rate = cohorte_numericas(dff, "cons.conf.idx")
bar_group_cohor(var_rate, x = "cons.conf.idx", y = "y", values = "Total")
```
